```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\BALAMURUGAN\OneDrive\Documents\Data - Base1.csv")

Age=df['Age']
Status=df['Marital status']

plt.bar(Age,Status)
plt.title("Age vs Avg Status")
plt.xlabel("Age")
plt.ylabel("Marital status")
plt.show()
```

```python
Age=df['Age']
Status=df['Marital status']

plt.scatter(Age,Status)
plt.title("Age vs Avg Status")
plt.xlabel("Age")
plt.ylabel("Marital status")
plt.show()
```

```python
Age = df['Age']
Status = df['Marital status']

plt.pie(Age.tail(5), labels=Status.tail(5))
plt.title("Last 5 Records - Age Distribution")
plt.show()
```
